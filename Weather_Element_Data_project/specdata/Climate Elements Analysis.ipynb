```r
f <- function(path, pollutant, id) {
    temp <- list.files(pattern="*.csv")
    for (i in temp) {
        path1 <- paste(path, i)
        path1 <- str_replace_all(string = path1, pattern = " ", repl= "")
#         print(path1)
        data <- read.csv(path1)
        
        if (unique(data$ID) == id) {
                myls <- c()
                for (i in data[pollutant]){
                        ii <- ifelse(is.na(i), "na", myls<- c(myls, i))
                    }

                bad <- is.na(myls)
                print(paste(pollutant, "Mean: ", mean(myls[!bad])))
              
                }

        else {

            next
        }
        
    }
    
}
                  
```

```r
f2 <- function(path, id ) {  
        matchFileName <- c()
        for (i in seq_len(length(id))){            
            if (nchar(trunc(i)) == 1){
               strng <- str_replace_all(string = paste("00", toString(i), ".csv"), pattern = " ", repl= "")
               matchFileName[i] <- strng
           }
           else if (nchar(trunc(i)) == 2){
               strng <- str_replace_all(string = paste("0", toString(i), ".csv"), pattern = " ", repl= "")
               matchFileName[i] <- strng
           }
           else {
               matchFileName[i] <- str_replace_all(string = paste(id[i], ".csv"), pattern = " ", repl= "")
           }                   
        }
        
        temp <- list.files(pattern="*.csv")
        IDs <- c()
        CaseCount <- c()
        
        for (j in matchFileName){
            if (is.element(j,temp)){
                
                path1 <- paste(path, j )
                path1 <- str_replace_all(string = path1, pattern = " ", repl= "")
                data <- read.csv(path1)
                CaseCount <- c(CaseCount, nrow(data))
                IDs <- c(IDs, unique(data$ID))
            }

            else {
                next
            }
        }
        print(data.frame(ids = IDs, Case_count = CaseCount))
}
```

```r
Corr <- function(path, threshold ) {
    
     getcor <- function(fname) {
#             print("-File name-")
#             print(paste(fname, ": "))
#             #print("-------------------the result is shown below----------------------")
            data <- read.csv(fname)
            new_data<- filter(data, (!is.na(sulfate) & !is.na(nitrate) ))
            corrlation<- cor(new_data[,2], new_data[,3])
            if (is.na(corrlation)){
                
                print("can not be calculated as outputed as 0")
                corrlation <- 0
            }
            
            else if (corrlation == 0){
                print("correlation is calculated as 0")
                next
            }
            bool_value <- as.numeric(corrlation) > as.numeric(threshold)
            if (bool_value) {
                print(corrlation)
            }

            else {
                print("Not enough for the threshold")
                
            }
          }
    temp <- list.files(path, pattern="*.csv")
    corr_result <- sapply(temp, getcor)
    corr_result <- unlist(corr_result[!sapply(corr_result, is.null)])
    return (corr_result)
}
```
