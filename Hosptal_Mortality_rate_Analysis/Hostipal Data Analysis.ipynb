Handling ties. It may occur that multiple hospitals have the same 30-day mortality rate for a given cause

```r
Ranking <- function(state, outcome, rAnk) {
    
    firstup <- function(x) {
    substr(x, 1, 1) <- toupper(substr(x, 1, 1))
    # substr('pneumonia', 1, 1) return "P"
    x
    }
    y <- list('HeartAttack' = 11, 
              'HeartFailure' = 17,
              'Pneumonia' = 23
             )
    library(stringi)
    outcome <- as.numeric(y[stri_replace_all_fixed(firstup(outcome), " ", "")])
    library(tidyverse)
    df <- read.csv("outcome-of-care-measures.csv", colClasses = "character")
    arrange1 <- arrange(df, State)
    Df <- select(arrange1, State, outcome, 2)
    newDf <- filter(Df, (State == state & !is.na(Df[,2]) & Df[,2] != "Not Available"))
    Sorted <- arrange(newDf, -desc(newDf[,2]))
    
    Order <- 1
    rank <- c()
    for (i in seq_len(length(Sorted[,2]))){
        val <- as.numeric(Sorted[,2][i]) == as.numeric(Sorted[,2][i+1])
        if(val & !is.na(as.numeric(Sorted[,2][i+1]))){
            rank[i] <- Order
            Order <- Order
            rank[i+1]<- Order   
        }
        else if (!val & !is.na(as.numeric(Sorted[,2][i+1]))) {
            rank[i] <- Order
            Order <- Order +1
            rank[i+1] <- Order    
        }
        else{   
            break
        }  
     }
    
    if (is.numeric(rAnk) & rAnk > Order | rAnk < 1){
        
        print("No Such Rank")
        
        
    }
    
    else{
    ## add rank col
    special <- list("best" = 1, "worst" = Order)
    if (is.numeric(rAnk)){
        Sorted_new <- Sorted %>% add_column(Rank=rank)
        x <- filter(Sorted_new, (Rank == rAnk)) 
    }
    else {
        Sorted_new <- Sorted %>% add_column(Rank=rank)
        x <- filter(Sorted_new, (Rank == as.numeric(special[tolower(rAnk)])))
    
    }
    
    x<- arrange(x, x[,3])
    select(x, Hospital.Name)
        
  }
}
```
